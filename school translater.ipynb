```python
import translate
from html import unescape

import time

import json
from googletrans import Translator
```

# load from article_list.json

```python
with open("article_list.json") as fp:
    aritcle_list=json.load(fp)["results"]

affils=set()

for article in aritcle_list:
    for aff in article["affils"]:
        affils.add(unescape(aff["name"]))
```

```python
d={"affils":[]}
for aff in affils:
    d["affils"].append(dict(name=aff,name_cn=""))
```

```python
with open("affils.json","w") as fp:
    json.dump(d,fp,indent=2)
```

# load from affils.json

```python
def reload():
    with open("affils.json","r") as fp:
        return json.load(fp)
```

```python
def save_reload(d):
    with open("affils.json","w") as fp:
        json.dump(d,fp,indent=2)
    return reload()
```

```python
def clear_trans():
    d=reload()
    for aff in d["affils"]:
        aff["name_cn"]=""
    save_reload(d)
```
